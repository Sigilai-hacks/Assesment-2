```python

df = pd.read_csv("C:/Users/kipng/OneDrive/Desktop/Sigilai Work Space/DATA SETS/amazon.csv")
df['discount_percentage'] = (
    df['discount_percentage']
    .str.replace('%', '', regex=True)
    .replace('', np.nan)  
    .astype(float)
)
df['rating'] = (
    df['rating']
    .str.replace(r'[^\d.]', '', regex=True)
    .replace('', np.nan) 
    .astype(float)
)
df_clean = df.dropna(subset=['discount_percentage', 'rating'])
plt.figure(figsize=(12, 8))
sns.scatterplot(
    x='discount_percentage',
    y='rating',
    data=df_clean,
    alpha=0.5,
    color='teal',
    edgecolor='black'
)
sns.regplot(x='discount_percentage', y='rating', data=df_clean, 
            scatter=False, color='red', line_kws={'linestyle':'--'})
plt.title('Discount Percentage vs Product Rating', fontsize=14)
plt.xlabel('Discount Percentage (%) →', fontsize=12)
plt.ylabel('← Product Rating (0-5)', fontsize=12)
plt.xlim(0, 100)
plt.ylim(0, 5)
plt.grid(True, alpha=0.3)

corr = df_clean['discount_percentage'].corr(df_clean['rating'])
plt.text(95, 0.5, f'Correlation: {corr:.2f}', ha='right', 
         bbox=dict(facecolor='white', alpha=0.8))

plt.show()
```

```python

df = pd.read_csv("C:/Users/kipng/OneDrive/Desktop/Sigilai Work Space/DATA SETS/amazon.csv")

df['rating_count'] = (
    df['rating_count']
    .str.replace(',', '', regex=True)
    .replace('', pd.NA)
    .pipe(pd.to_numeric, errors='coerce')
)
category_sales = (
    df.groupby('category', as_index=False)['rating_count']
    .sum()
    .sort_values('rating_count', ascending=False)
)
top_categories = category_sales.head(10)
other = pd.DataFrame({
    'category': ['Others'],
    'rating_count': [category_sales['rating_count'][10:].sum()]
})
final_data = pd.concat([top_categories, other], ignore_index=True)
plt.figure(figsize=(12, 8))
colors = plt.cm.tab20.colors
explode = [0.1] + [0]*(len(final_data)-1)  

plt.pie(final_data['rating_count'],
        labels=final_data['category'],
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,
        explode=explode,
        wedgeprops={'edgecolor': 'black', 'linewidth': 0.5},
        pctdistance=0.85)

centre_circle = plt.Circle((0,0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)
plt.title('Top-Selling Product Categories\n(by Review Count)', fontsize=16, pad=20)
plt.axis('equal')
plt.legend(final_data['category'],
           title="Categories",
           loc="center left",
           bbox_to_anchor=(1, 0, 0.5, 1))

plt.tight_layout()
plt.show()
```

```python

df = pd.read_csv("C:/Users/kipng/OneDrive/Desktop/Sigilai Work Space/DATA SETS/amazon.csv")
def clean_numeric(col):
    return (
        col.astype(str)
        .str.replace('[^\d.]', '', regex=True)  # Remove non-numeric characters
        .replace('', np.nan)
        .astype(float)
    )

numerical_cols = ['discounted_price', 'actual_price', 'discount_percentage', 
                 'rating', 'rating_count']

df_clean = df[numerical_cols].apply(clean_numeric)

corr_matrix = df_clean.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, 
            annot=True, 
            cmap='coolwarm', 
            vmin=-1, 
            vmax=1, 
            linewidths=0.5,
            annot_kws={'size': 12},
            mask=np.triu(np.ones_like(corr_matrix, dtype=bool)))

plt.title('Correlation Heatmap of Numerical Features', fontsize=16, pad=20)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
